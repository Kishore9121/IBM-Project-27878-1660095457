## ***5.Outliers Detection***

```python
plt.figure(figsize=(20,10))
df.boxplot()
plt.show()
```

```python
def outliers(col):
  Q3,Q1=np.nanpercentile(col,[75,25])
  IQR=Q3-Q1
  upper_limit=IQR+1.5*Q3
  lower_limit=IQR-1.5*Q1
  return upper_limit,lower_limit
```

## ***Encoding***

```python
df_new=pd.get_dummies(df,columns=['Sex'])
```

```python
df_new.head()
```

```python
df_new.shape
```

## ***7.Splitting into dependent and independent variables***

```python
Independent=df_new.drop(['Age'],axis=1)
Dependent=df_new['Age']
```

```python
Independent.head()
```

```python
Dependent.head()
```

## ***8.Scaling the independent features***

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=pd.DataFrame(sc.fit_transform(Independent),columns=Independent.columns)
```

```python
X.head()
```

## ***9.Splitting Dataset Into Training And Testing***

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Dependent,test_size=0.3,random_state=0)
```

```python
X_train.reset_index().head()
```

```python
X_train.shape
```

```python
X_test.shape
```

## ***10.Model Building***

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
```

```python
import xgboost as xg
xgb=xg.XGBRegressor()
model1=xgb.fit(X_train,Y_train)
```

## ***11.Training The Model***

```python
model=rfr.fit(X_train,Y_train)
```

## ***12.Testing The Model***

```python
y_pred_xg=model1.predict(X_test)
```

```python
y_pred=model.predict(X_test)
```

## ***Accuracy***

```python
from sklearn import metrics
print("the accuracy of xgboost regressor model:",metrics.r2_score(Y_test,y_pred_xg))
print("the accuracy of random fore model:",metrics.r2_score(Y_test,y_pred_xg))
```
